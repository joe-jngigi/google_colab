```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original_data = pd.read_csv("https://raw.githubusercontent.com/ZUCourses/SIA-Public/main/Data%20Sets/CAE_GPA.csv")

# Select a random sample of 100 students
df=original_data.sample(n=100)

# Save to CSV file
df.to_csv(r'/FilName.csv')
print (df)
```

```python
# Create a histogram and generate descriptive statistics for each quantitative variable
describe_quatity = ['sex', 'CAE_v', 'CAE_m', 'CAE_sum', 'hs_gpa', 'fy_gpa']

for desc_variable in describe_quatity:
    # Create a histogram
    plt.hist(df[desc_variable], bins='auto')
    plt.title(desc_variable)
    plt.xlabel(desc_variable)
    plt.ylabel("Frequency")
    plt.show()
    
    # Descriptive statistics
    print(f"Descriptive statistics for {desc_variable}: \n{ df[desc_variable].describe()}")
```

```python
# Grouped box plot
sns.boxplot(x='sex', y='hs_gpa', data=df)
plt.title("GPA by Sex")
plt.xlabel("sex")
plt.ylabel("High School GPA")
plt.show()

# Five-number summaries
male_summary = df.loc[df['sex'] == 1]['hs_gpa'].describe()
female_summary = df.loc[df['sex'] == 2]['hs_gpa'].describe()

print(f"Male GPA Summary: \n{male_summary}")
print("\n")
print(f"Female l GPA Summary: \n{female_summary}")
```

```python
# Select only male and female students
ndf = original_data.loc[original_data['sex'].isin([1, 2])]

# Grouped box plot
sns.boxplot(x='sex', y='fy_gpa', data=ndf)
plt.title("GPA by Gender")
plt.xlabel("sex")
plt.ylabel("Freshman GPA")
plt.show()

# Five-number summary
male_summary = ndf.loc[ndf['sex'] == 1]['fy_gpa'].describe()
female_summary = ndf.loc[ndf['sex'] == 2]['fy_gpa'].describe()

print(f"Male Freshman GPA Summary: \n{male_summary}")
print("\n")
print(f"Female Freshman GPA Summary: \n{female_summary}")
```
