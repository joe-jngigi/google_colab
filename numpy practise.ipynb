```python
# Le us say we want to get a specific element
new_darray = np.array([
    [12, 23, 24, 45, 21, 2, 34, 21],
    [13, 33, 24, 5, 2, 2, 34.3, 41]
], dtype = np.int32)
# print array
print(new_darray)
# get shape of array
print(new_darray.shape)
# Get the dimensions
print(new_darray.ndim)

# Get a specific element, rows and columns
print(f'The number at the seventh column of second row is {new_darray[1, 6]}')

# To get everything in a given row, we use =>similar to a list )(all rows = :)
print(f'The first row has the following {new_darray[0, :]}')

# Get a specific column 
print(f'Get a column {new_darray[:, 3]}')

# Lets say I want 23, 45, 2. So it is specify the row, get the first and the last index 1 & 7, then jump two steps ahead
print(f'Steps {new_darray[0, 1:7:2]}')

# I can also change an element at a certain point
print('Change Elemnt ant the second row in the fifth column')
new_darray[1, 4] = 24.7
print(new_darray)

# I can also change an entire row or column
new_darray[:, 5] = 15
print(new_darray)
# Entire row
new_darray[1, :] = 0
print(new_darray)
##### Note: it does not return decimals because I specified/ changed the data type to an integer
```
