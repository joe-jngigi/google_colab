```python
# Le us say we want to get a specific element
new_darray = np.array([
    [12, 23, 24, 45, 21, 2, 34, 21],
    [13, 33, 24, 5, 2, 2, 34.3, 41]
], dtype = np.int32)
# print array
print(new_darray)
# get shape of array
print(new_darray.shape)
# Get the dimensions
print(new_darray.ndim)

# Get a specific element, rows and columns
print(f'The number at the seventh column of second row is {new_darray[1, 6]}')

# To get everything in a given row, we use =>similar to a list
print(f'The first row has the following {new_darray[0, :]}')

# Note: it does not return decimals because I specified/ changetr
```
